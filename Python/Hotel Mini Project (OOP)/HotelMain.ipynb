```python
from hotel_booking import Hotel, User
```

```python
def main():
    hotel = Hotel()
    hotel.addRoom(101, roomType='standard')
    hotel.addRoom(102, roomType='premium')
    hotel.addRoom(103, roomType='standard')
    hotel.addRoom(201, roomType='premium')
    hotel.addRoom(202, roomType='standard')
    hotel.addRoom(203, roomType='premium')
    
    user = User("neb", "zeynep123", [
        {"bank_name": "A Bank","balance": 1250},
        {"bank_name": "B Bank","balance": 2350}
    ])
    
    print("\n======== Welcome to the Marbella Hotel! ========")
    
    while True:
        print("1. Show Rooms")
        print("2. Book Room")
        print("3. Make Payment")
        print("4. Exit")
        
        option = input("Enter your option please (1-4): ")
        
        if option == "1":
            hotel.displayRooms()
        elif option == "2":
            roomNum = int(input("Enter the room number: "))
            numNights = int(input("Enter the number of nights: "))
            hotel.bookRoom(roomNum, numNights)
        elif option == "3":
            hotel.makePayment(user)
        elif option == "4":
            print("Thank you for using the Marbella Hotel. We hope to see you again!")
            break
        else:
            print("Invalid choice. Please enter valid choice (1-4).")
```

```python
if __name__ == "__main__":
    main()
```
